## Week 3 - Part 1

```python
# extract data from the wikipedia table

from pandas.io.html import read_html

page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

table = read_html(page, attrs ={"class":"wikitable"})

print("Extracted {num} wikitables".format(num = len(table)))
```

```python
# view extracted data to confirm it's the right data

table[0].head()
```

```python
# convert the extracted data into a dataframe, labelled as df

df = table[0]
```

```python
# view the first five columns of the dataframe to confirm that it's correct

df.head()
```

```python
# replace the "Not assigned" in the Borough column to "NaN" so that those rows can be dropped

df.replace('Not assigned','NaN', inplace = True)
```

```python
# drop the rows that are "NaN" in the Borough column

df.dropna(axis = 0, inplace = True)
```

```python
# view the first 5 columns of the dataframe

df.head()
```

```python
# reset the index after dropping the rows

df.reset_index(drop = True)
```

```python
# showed the shape of the resulting data frame as per the instructions

df.shape
```
